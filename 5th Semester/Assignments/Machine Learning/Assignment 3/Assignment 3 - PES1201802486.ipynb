```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('stock_infosys.csv')
```

```python
df.describe()
```

```python
df.head()
```

```python
y=df['Close']
x=df[['Open','High','Low','Vol-Million','Change %']]
y.shape
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size = 0.3, random_state = 40)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.tree import DecisionTreeRegressor

model = DecisionTreeRegressor(criterion='mse',max_depth=5,min_samples_split=20)
```

```python
model.fit(x_train,y_train)
```

```python
mod_pred=model.predict(x_train)

from sklearn.metrics import mean_squared_error,r2_score
from math import sqrt

rms=sqrt(mean_squared_error(y_train,mod_pred))
print(rms)
```

```python
r2=r2_score(y_train,mod_pred)
print(r2)
```

```python
test_pred=model.predict(x_test)
rms=sqrt(mean_squared_error(y_test,test_pred))
print(rms)
r2=r2_score(y_test,test_pred)
print(r2)
```

```python
from sklearn.tree import export_graphviz
from IPython.display import Image
import pydot 
```

```python
from sklearn.externals.six import StringIO
data=StringIO()
export_graphviz(model,out_file=data,filled=True,max_depth=5,feature_names=x.columns)
graph=pydot.graph_from_dot_data(data.getvalue())
Image(graph[0].create_png())
```

```python
from sklearn.ensemble import BaggingRegressor,RandomForestRegressor
```

```python
bagm=BaggingRegressor(n_estimators=6)
bagm.fit(x_train,y_train)
pred=bagm.predict(x_test)
```

```python
rms=sqrt(mean_squared_error(y_test,pred))
print(rms)
```

```python
r2=r2_score(y_test,pred)
print(r2)
```

```python
rfm=RandomForestRegressor(n_estimators=6)
rfm.fit(x_train,y_train)
pred=rfm.predict(x_test)
rms=sqrt(mean_squared_error(y_test,pred))
print(rms)
r2=r2_score(y_test,pred)
print(r2)
```

```python
from sklearn.ensemble import AdaBoostRegressor
adam=AdaBoostRegressor(n_estimators=6)
adam.fit(x_train,y_train)
pred=adam.predict(x_test)
rms=sqrt(mean_squared_error(y_test,pred))
print(rms)
r2=r2_score(y_test,pred)
print(r2)
```
