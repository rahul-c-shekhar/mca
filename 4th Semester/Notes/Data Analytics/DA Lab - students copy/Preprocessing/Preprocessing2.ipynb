```python
import pandas as pd
import numpy as np
from sklearn import linear_model
from math import sqrt
from sklearn.metrics import mean_squared_error
```

```python
df=pd.read_csv('Housing.csv')
df.head()
```

```python
df.columns
```

```python
df.tail()
```

```python
#what is that you want to predict
y=df['price']
print(type(y))
y=y.values
y=y.reshape(len(y),1)
print(y)
```

```python
degree=[1,2,3]
x=df['lotsize']
lt=pd.cut(x,3,labels=degree)
print(lt)
```

```python
df['lt']=lt
df['lt']=df['lt'].astype('category')
x=df[['lt','bedrooms']]
#x=x.values
#x=x.reshape(len(x),1)
print(x)
```

```python
x_train=x[:430]
x_test=x[431:]
y_train=y[:430]
y_test=y[431:]
regr=linear_model.LinearRegression()
regr.fit(x_train,y_train)
pred=regr.predict(x_test)
print(pred)
```

```python
rms=sqrt(mean_squared_error(y_test,pred))
print(rms)
```
